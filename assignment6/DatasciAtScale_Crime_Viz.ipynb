```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
%matplotlib inline
```

```python
sanfran = pd.read_csv('sanfrancisco_incidents_summer_2014.csv')
seatt = pd.read_csv('seattle_incidents_summer_2014.csv')
```

```python
seatt.head()
```

```python
sanfran.head()
```

```python
seatt.columns
```

```python
sanfran.columns
```

```python
seatt['timestamp'] = seatt['Occurred Date or Date Range Start']
# Break up the date into hour, weekday, month, year
seatt_datetime = pd.DatetimeIndex(seatt['timestamp'])
```

```python
seatt['crime_hour'] = seatt_datetime.hour
seatt['crime_weekday'] = seatt_datetime.weekday
seatt['crime_month'] = seatt_datetime.month
seatt['crime_year'] = seatt_datetime.year
```

```python
sanfran['Date'].head()
```

```python
sanfran['Time'].head()
```

```python
# We need to combine these columns
sanfran['timestamp'] = pd.to_datetime(sanfran['Date'] + ' ' + sanfran['Time'])
sanfran_datetime = pd.DatetimeIndex(sanfran['timestamp'])
```

```python
sanfran['crime_hour'] = sanfran_datetime.hour
sanfran['crime_weekday'] = sanfran_datetime.weekday
sanfran['crime_month'] = sanfran_datetime.month
sanfran['crime_year'] = sanfran_datetime.year
```

```python
# There are certain crimes which do not occure very often, so I will remove them.
sanfran_crime_numbers = sanfran['Category'].value_counts()
sanfran_crime_numbers
```

```python
sanfran_crimes = sanfran['Category'].value_counts().keys()[:13]
sanfran_crimes
```

```python
np.arange(len(sanfran_crimes[:13]))
```

```python
# I'll cut these off at Prostitution
seatt_crime_numbers = seatt['Summarized Offense Description'].value_counts()
seatt_crime_numbers
```

```python
seatt_crimes = seatt_crime_numbers.keys()[:20]
seatt_crimes
```

```python
np.arange(len(seatt_crimes[:20]))
```

```python
fig = plt.figure(figsize=(6,4))
ax = plt.gca()
plt.subplots_adjust(bottom=0.17,left=0.17,top=0.96,right=0.96)

# Change the axis units to CMU Serif
plt.setp(ax.get_ymajorticklabels(),family='serif',fontsize=18)
plt.setp(ax.get_xmajorticklabels(),family='serif',fontsize=18)

ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')

ax.xaxis.set_ticks_position('bottom')
ax.yaxis.set_ticks_position('left')

# Turn on the plot grid and set appropriate linestyle and color
ax.grid(True,linestyle=':',color='0.75')
ax.set_axisbelow(True)

plt.barh(np.arange(len(sanfran_crimes[:13])),sanfran_crime_numbers.get_values()[:13])
# axs[1].barh(np.arange(len(sanfran_crimes[:13])),sanfran_crime_numbers.get_values()[:13])
# Define the X and Y axis labels
plt.xlabel('Frequency of Crimes',family='serif',fontsize=22,weight='bold',labelpad=5)
plt.ylabel('Description',family='serif',fontsize=22,weight='bold',labelpad=10)
plt.yticks(np.arange(len(sanfran_crimes)), map(lambda x: x.title(),
                                               sanfran_crimes), fontsize= 14)
plt.title('San Francisco Crimes',fontsize=22,weight='bold')

# Adjust the page layout filling the page using the new tight_layout command
plt.tight_layout(pad=0.5)

# save the figure as a high-res pdf in the current folder
# plt.savefig('vibration_data.pdf',dpi=600)

fig.set_size_inches(9,6) # Resize the figure for better display in the notebook
```

```python
# Now I am going to look at the crimes of each city for the hour, day, and month
sanfran_all_crime_hour = sanfran.groupby('crime_hour').size()
seatt_all_crime_hour = seatt.groupby('crime_hour').size()

sanfran_all_crime_weekday = sanfran.groupby('crime_weekday').size()
seatt_all_crime_weekday = seatt.groupby('crime_weekday').size()

sanfran_all_crime_month = sanfran.groupby('crime_month').size()
seatt_all_crime_month = seatt.groupby('crime_month').size()

sanfran_all_crime_year = sanfran.groupby('crime_year').size()
seatt_all_crime_year = seatt.groupby('crime_year').size()
```
